```python
from sklearn.cluster import KMeans
#Train our model
km =KMeans(n_clusters=2)
clusters=km.fit_predict(d.iloc[:,-3:])
```

```python
d["label"] = clusters
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
fig = plt.figure(figsize=(40,30))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(d.Age[d.label == 0], d["Annual Income (k$)"][d.label == 0], d["Spending Score (1-100)"][d.label == 0], c='blue', s=60)
ax.scatter(d.Age[d.label == 1], d["Annual Income (k$)"][d.label == 1], d["Spending Score (1-100)"][d.label == 1], c='red', s=60)

ax.view_init(30, 185)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k,init="k-means++")
  kmeans.fit(d.iloc[:,-3:])
  wcss.append(kmeans.inertia_)

print(wcss)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="*")
plt.xlabel("k values")
plt.xticks(np.arange(1,11,1))
plt.ylabel("Variance - Error (WCSS)")
plt.show()
```
